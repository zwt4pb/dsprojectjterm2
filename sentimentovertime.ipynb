```python
import pandas as pd
import matplotlib.pyplot as plt


file_path_new = '/root/justin_bieber.csv'
data_new = pd.read_csv(file_path_new)

data_new['Release Date'] = pd.to_datetime(data_new['Release Date'], errors='coerce')

data_new = data_new.sort_values(by='Release Date')

color_map = {'Positive': 'green', 'Negative': 'red', 'Neutral': 'blue'}
data_new['Color'] = data_new['Sentiment Category'].map(color_map)

plt.figure(figsize=(14, 7))
plt.scatter(data_new['Release Date'], data_new['Sentiment Score'], c=data_new['Color'], alpha=0.8, edgecolor='k')
plt.title('Sentiment Scores of Justin Bieber Songs Over Time', fontsize=16)
plt.xlabel('Release Date', fontsize=12)
plt.ylabel('Sentiment Score', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

justin_bieber_data = [
    {"artist": "Justin Bieber", "type": "Album", "name": "My World 2.0", "date": "2010-03-19"},
    {"artist": "Justin Bieber", "type": "Album", "name": "Believe", "date": "2012-06-15"},
    {"artist": "Justin Bieber", "type": "Album", "name": "Under the Mistletoe", "date": "2011-11-01"},
    {"artist": "Justin Bieber", "type": "Album", "name": "Purpose", "date": "2015-11-13"},
    {"artist": "Justin Bieber", "type": "Album", "name": "Changes", "date": "2020-02-14"},
    {"artist": "Justin Bieber", "type": "Album", "name": "Justice", "date": "2021-03-19"},
    {"artist": "Justin Bieber", "type": "Event", "name": "Relationship with Selena Gomez became public", "date": "2011-01-01"},
    {"artist": "Justin Bieber", "type": "Event", "name": "Arrested for DUI in Miami", "date": "2014-01-23"},
    {"artist": "Justin Bieber", "type": "Event", "name": "Canceled Purpose World Tour due to depression and anxiety", "date": "2017-07-24"},
    {"artist": "Justin Bieber", "type": "Event", "name": "Social media hiatus after backlash over relationship with Sofia Richie", "date": "2016-08-15"},
    {"artist": "Justin Bieber", "type": "Event", "name": "Married Hailey Baldwin in courthouse ceremony", "date": "2018-09-13"},
    {"artist": "Justin Bieber", "type": "Event", "name": "Opened up about Lyme disease diagnosis", "date": "2020-01-08"},
    {"artist": "Justin Bieber", "type": "Event", "name": "Took a break from touring due to Ramsay Hunt syndrome", "date": "2023-06-01"}
]

justin_bieber_df = pd.DataFrame(justin_bieber_data)
justin_bieber_df['date'] = pd.to_datetime(justin_bieber_df['date'])
color_map_jb = {'Album': 'blue', 'Event': 'orange'}
justin_bieber_df['Color'] = justin_bieber_df['type'].map(color_map_jb)

data_new['Release Date'] = pd.to_datetime(data_new['Release Date'])

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=data_new['Release Date'],
    y=data_new['Sentiment Score'],
    mode='markers',
    marker=dict(color=data_new['Color'], size=10, opacity=0.8),
    name="Sentiment Scores"
))

fig.add_trace(go.Scatter(
    x=justin_bieber_df.loc[justin_bieber_df['type'] == 'Album', 'date'],
    y=[1.5] * len(justin_bieber_df[justin_bieber_df['type'] == 'Album']),
    mode='markers',
    marker=dict(size=10, color='blue', symbol='diamond'),
    hovertext=justin_bieber_df.loc[justin_bieber_df['type'] == 'Album', 'name'],
    name="Justin Bieber Albums"
))

fig.add_trace(go.Scatter(
    x=justin_bieber_df.loc[justin_bieber_df['type'] == 'Event', 'date'],
    y=[1.2] * len(justin_bieber_df[justin_bieber_df['type'] == 'Event']),
    mode='markers',
    marker=dict(size=10, color='orange', symbol='circle'),
    hovertext=justin_bieber_df.loc[justin_bieber_df['type'] == 'Event', 'name'],
    name="Justin Bieber Events"
))

fig.update_layout(
    title="Overlay of Justin Bieber Sentiment Scores, Albums, and Events",
    xaxis_title="Date",
    yaxis_title="Sentiment Score",
    height=600,
    xaxis=dict(rangeslider=dict(visible=True)),
    showlegend=True
)

fig.show()
```
